```python
# Mudando o tipo de uma coluna
df = df.with_columns(
    pl.col("data_compra").str.to_date().alias("data_compra")
)
```

```python
# estatisticas
df['vendas'].mean()
```

```python
# olhando a quantidade de nulos
df.null_count().mean()
```

```python
# Preenchendo os valores faltantes
df = df.fill_nan(-9)
```

```python
# Exluindo os valores nulos
df = df.drop_nans()
```

```python
# Removendo linhas duplicadas
df = df.unique()  # df.unique(subset=["coluna"], keep="first")
```

```python
# Criando uma coluna
df = df.with_columns(
    (pl.col('vendas') * pl.col('preco')).alias('total')
)
df['total']
```

```python
# Passando uma função em uma coluna
def foobar(value: float) -> int:
    return value / 100

df.with_columns(
    pl.col("preco").map_elements(foobar, return_dtype=pl.Float64).alias('temp')
)['temp']
```
