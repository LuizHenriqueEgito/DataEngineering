```python
# Renomeando o nome das colunas
rename_dict = {
    'sales': 'vendas',
    'customer_id': 'id_cliente',
    'customer_age': 'idade_cliente',
    'purchase_date': 'data_compra',
    'product': 'produto',
    'last_purchase_date': 'data_ultima_compra',
    'price': 'preco',
    'seller_id': 'id_vendedor'
}
df = df.rename(columns=rename_dict)
df.columns
```

```python
# Seleciona uma coluna especifica
df.produto
```

```python
# Mudando o tipo de uma coluna
df['data_compra'] = pd.to_datetime(df['data_compra'])
```

```python
# Faz um filtro para vendas após 2023-09-09
df.query('data_compra >= "2023-09-09"')
```

```python
# Contando as vendas por vendedor
df.id_vendedor.value_counts(normalize=True)
```

```python
# Selecionando 3 colunas, id_cliente, produto, preço
df[['id_cliente', 'produto', 'preco']]
```

```python
df.loc[:, ['id_cliente', 'produto', 'preco']]
```

```python
# Selecionando apenas as colunas que comecem com id
df.filter(regex="^id", axis=1)
```

```python
df[[col for col in df.columns if col.startswith("id")]]
```

```python
# estatisticas
df.vendas.mean()
```

```python
# Olhando a quantidade de nulos
df.isna().mean()
```

```python
# Preenchendo os valores faltantes
df = df.fillna(-9)
```

```python
# Exluindo os valores nulos
df = df.dropna()
```

```python
# Removendo linhas duplicadas
df = df.drop_duplicates()
```

```python
# Criando uma coluna
df['total'] = df.vendas * df.preco
df.total
```

```python
# Passando uma função em uma coluna
def foobar(value: float) -> int:
    return value / 100

df['preco'].apply(foobar)
```

```python
df.groupby(by='produto')['vendas'].mean()
```
